```python
#Import label encoder
from sklearn import preprocessing
#label_encoder object knows how to understand word labels
label_encoder = preprocessing.LabelEncoder()
#Encode labels in column Sex and Embarked
dataset['Sex']= label_encoder.fit_transform(dataset['Sex'])
dataset['Embarked']=label_encoder.fit_transform(dataset['Embarked'])
#Setting the value for dependent and independent variables
X = dataset.drop('Survived', 1)
y = dataset.Survived
print("Features (X): \n",X)
print("Target (y): \n",y)
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
random_state=2)
```

```python
#Fitting the Logistic Regression model
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

```python
#Prediction of test set
y_pred = lr_model.predict(X_test)
#Predicted values
y_pred
```

```python
#Actual value and the predicted value
diff = pd.DataFrame({'Actual value': y_test, 'Predicted value':y_pred})
```

```python
diff.head()
```

```python
#Confusion matrix and classification report
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(matrix, annot=True, fmt="d")
plt.title('Confusion M trix')
plt.xlabel('Predicted')
plt.ylabel('True')
print(classification_report(y_test, y_pred))
```
