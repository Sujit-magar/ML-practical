```python
# Import dataset
dataset = pd.read_csv("/content/drive/MyDrive/Social_Network_Ads.csv")
dataset.head()
```

```python
# Slice input and output
X = dataset.iloc[:, 2:4]
y = dataset.iloc[: , -1]
```

```python
print("Input Features\n")
X
```

```python
print("Output/Target Feature: \n")
y
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print("Training Data(Input): \n")
X_train
```

```python
print("Training Data(Output): \n")
y_train
```

```python
print("Test Data(Input): \n")
X_test
```

```python
print("Test Data(Output): \n")
y_test
```

```python
# Feeature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Training the Naive Bayes Model on the Training Set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
print("Predicted Values: \n")
y_pred
```

```python
print("Actual Values: \n")
y_test
```

```python
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix: \n", cm)
```

```python
ac = accuracy_score(y_test, y_pred)
print("Accuracy of the model: ", ac*100, "%")
```
